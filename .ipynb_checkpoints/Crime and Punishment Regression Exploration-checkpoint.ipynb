## Exploring the relationship between crime and punishment using Linear Regression 
\- Writtn by Leonard Phillips.

This report investigates whether there is a relationship between the crime rate of society and the probability of imprisonment. The dependent variable 'Crime' represents the number of offenses per 1,000,000 population while the independent variable 'ProbPrison' represents the ratio of the number of people committed to prison to the number of offenses.

The dataset I will be using in this report has data collected during the year 1960.

#### Importing the data
As the file we are reading the data from is a SPSS file, a .sav file reader must be imported first. The package I have imported can be found <a href="https://github.com/Roche/pyreadstat">here.</a>

```python
import pyreadstat
import pandas as pd

df, meta = pyreadstat.read_sav("Punishment and Crime.sav")
crimAndPunDF = df[['Crime', 'ProbPrison']]
```

```python
import statsmodels.formula.api as smf
def regress_model(df,target,predictor):
    poly_1_formula = target + ' ~ 1 + ' + predictor
    poly_2_formula = target + ' ~ 1 + ' + predictor + ' + I(' + predictor + ' ** 2.0)'
    poly_3_formula = target + ' ~ 1 + ' + predictor + ' + I(' + predictor + ' ** 2.0) + I('+ predictor + ' ** 3.0)'    
    
    # 1-st order polynomial (Straight line)
    poly_1 = smf.ols(formula=poly_1_formula, data=df.dropna()).fit()

    # 2-nd order polynomial (Quadratic)
    poly_2 = smf.ols(formula=poly_2_formula, data=df.dropna()).fit()

    # 3-rd order polynomial (Cubic)
    poly_3 = smf.ols(formula=poly_3_formula, data=df.dropna()).fit()
    
    return poly_1, poly_2, poly_3
```

```python
import scipy

def linregress_CIs(xd,yd,conf=0.95):
    """Linear regression CIs """
    alpha=1.-conf   # significance
    n = xd.size   # data sample size
    x = np.linspace(xd.min(),xd.max(),1000)
        
    # Predicted values from fitted model:
    a, b, r, p, err = scipy.stats.linregress(xd,yd)
    y = a*x+b
    
    sd = 1./(n-2.)*np.sum((yd-a*xd-b)**2)
    sd = np.sqrt(sd)
    sxd = np.sum((xd-xd.mean())**2) #SS total
    sx  = (x-xd.mean())**2 # variance of each x
    
    # quantile of student's t distribution for p=1-alpha/2
    q = scipy.stats.t.ppf(1.-alpha/2, n-2)
    
    # get the upper and lower CI:
    #print 'sx/sxd:', sx/sxd
    dy = q*sd*np.sqrt( 1./n + sx/sxd )
    
    yl = y-dy
    yu = y+dy
    
    return yl,yu,x
```

```python
import matplotlib.pyplot as plt
import numpy as np
def visualize_regression(df, target, predictor):
    
    fig = plt.figure(figsize=[20,20])
    fig.subplots_adjust(hspace=0.4, wspace=0.2)
    
    
    subplot_index = 1
    r_squared = 0
    
    ax = plt.subplot('32'+str(subplot_index))
    subplot_index += 1

    ax.scatter(df[predictor], df[target], alpha=0.5)
    ax.set_xlabel(predictor)
    ax.set_ylabel(target)

    poly_1, poly_2, poly_3 = regress_model(df,target,predictor)
    x = pd.DataFrame({predictor: np.linspace(df[predictor].min(), df[predictor].max(), len(df[predictor]))})
    
    yl,yu,xd = linregress_CIs(df[predictor].as_matrix(),df[target].as_matrix(), .95)

    ax.plot(x, poly_1.predict(x), 'y-', label='Poly n=1 $R^2$=%.2f' % poly_1.rsquared)
    ax.plot(x, poly_2.predict(x), 'g-', label='Poly n=2 $R^2$=%.2f' % poly_2.rsquared)
    ax.plot(x, poly_3.predict(x), 'r-', label='Poly n=3 $R^2$=%.2f' % poly_3.rsquared)
    
    ax.fill_between(xd, yl, yu, alpha=0.3, facecolor='black', edgecolor='none')

    ax.legend(loc='best', prop={'size':'small'})
        
    plt.show()
    return poly_1, poly_2, poly_3
    
```

```python
poly_1, poly_2, poly_3 = visualize_regression(crimAndPunDF, 'Crime', 'ProbPrison')
```

A visual examination of the graph shows that both the third and second-order polynomials fit the data better than a straight-line model. This is represented by the 0.25 and 0.24 R squared values respectively.
